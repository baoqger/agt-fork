```python
from agentmesh.governance import GovernanceDenied, govern

def web_search(query: str) -> str:
    return f"Results for: {query}"

def delete_file(path: str) -> str:
    return f"Deleted: {path}"

safe_search = govern(
    lambda action, query: web_search(query),
    policy="policy.yaml",
)
safe_delete = govern(
    lambda action, path: delete_file(path),
    policy="policy.yaml",
)

# Option 2: original tool functions stay unchanged, but callers must pass action.
print(safe_search(action="web_search", query="AI governance news"))

# This should now raise GovernanceDenied because policy.yaml denies delete_file.
try:
    print(safe_delete(action="delete_file", path="/etc/passwd"))
except GovernanceDenied as exc:
    print(f"Blocked: {exc}")
```
